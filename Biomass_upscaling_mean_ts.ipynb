```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

class TS_dense(nn.Module):
    
    """(convolution => [BN] => ReLU) * 2"""
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.fc1 = nn.Sequential(
            nn.Linear(in_channels, 10000),
            nn.Tanh(),
            nn.BatchNorm2d(6),
            nn.Dropout(.5),
            
            nn.Linear(10000, 2000),
            nn.Tanh(),
            nn.BatchNorm2d(6),
            nn.Dropout(.4),
            
            nn.Linear(2000, 1000),
            nn.Tanh(),
            nn.BatchNorm2d(6),
            nn.Dropout(.2),
            
            nn.Linear(1000, 400),
            nn.Tanh(),
            nn.BatchNorm2d(6),
            
            nn.Linear(400, 40),
            nn.Tanh(),
            nn.Linear(40, 7),
        )


    def forward(self, x):
        x = x.unsqueeze(dim=0)
        x=self.fc1(x)
        x = x.squeeze(dim=0)
        return x
```

```python
def run_genetic_algorithm(population_size, num_generations, mutation_rate):
    # Initialize population
    population = [generate_individual() for _ in range(population_size)]
        # Evaluate fitness
    for generation in range(num_generations):
    # Evaluate fitness
        fitnesses = [compute_fitness(individual) for individual in population]
        
        # Elitism: Carry over the best individuals
        sorted_population = [x for _, x in sorted(zip(fitnesses, population))]
        new_population = sorted_population[:elitism_size]
    
        while len(new_population) < population_size:
            parent1, parent2 = select_parents(population, fitnesses)
            child1 = crossover(parent1, parent2)
            child2 = crossover(parent1, parent2)
            new_population.extend([mutate(child1), mutate(child2)])
    
        population = new_population
    
        # Print the best fitness every 100 generations
        if generation % 100 == 0:
            best_individual = population[0]
            best_fitness = compute_fitness(best_individual)
            print(f"Generation {generation}: Best MSE = {best_fitness}")
    
        # Final best solution
        best_individual = min(population, key=compute_fitness)
        best_fitness = compute_fitness(best_individual)
    
        print("Final Best individual:", best_individual)
        print("Final Best fitness (MSE):", best_fitness)

# Run the algorithm with one set of parameters
run_genetic_algorithm(population_size=100, num_generations=1000, mutation_rate=0.1)

# Reset and run the algorithm with a different set of parameters
#run_genetic_algorithm(population_size=200, num_generations=500, mutation_rate=0.05)
```

```python
import numpy as np

def calculate_divergence(vx, vy, dx, dy):
    """
    Calculate the divergence of a 2D velocity field using a forward-looking finite difference scheme.
    
    Parameters:
    vx: 2D array of velocity in the x-direction
    vy: 2D array of velocity in the y-direction
    dx: Grid spacing in the x-direction
    dy: Grid spacing in the y-direction
    
    Returns:
    divergence: 2D array of divergence values
    """
    # Get the dimensions of the velocity field
    rows, cols = vx.shape

    # Initialize an array to store divergence
    divergence = np.zeros((rows, cols))

    # Calculate divergence using forward differences
    # For the edges, we use backward differences to avoid index out of range
    for i in range(rows):
        for j in range(cols):
            dvx_dx = (vx[i, min(j+1, cols-1)] - vx[i, j]) / dx
            dvy_dy = (vy[min(i+1, rows-1), j] - vy[i, j]) / dy
            divergence[i, j] = dvx_dx + dvy_dy

    return divergence

# Example usage with dummy data
# Create sample velocity fields
#vx = outputs[100,:,:,0].permute(1,0).cpu().detach().numpy()

#vy = outputs[100,:,:,1].permute(1,0).cpu().detach().numpy()


# Assume uniform grid spacing
#dx = 1  # x-spacing
#dy = 1  # y-spacing

# Calculate divergence
#divergence = calculate_divergence(vx, vy, dx, dy)
#divergence
```
